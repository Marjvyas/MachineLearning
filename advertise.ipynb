```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
df=pd.read_csv('/content/sample_data/Advertising.csv')
df
df.drop(columns=['Unnamed: 0'], inplace=True)
```

```python
df.info()
```

```python
df.describe()
```

```python
X=df[['TV', 'radio', 'newspaper']]
y=df['sales']
```

```python
X_train, X_test, y_train, y_test=train_test_split(X, y, test_size=0.3, random_state=1)
```

```python
model=LinearRegression()
model.fit(X_train, y_train)
```

```python
y_pred=model.predict(X_test)
```

```python
for column in X_test.columns:
    plt.figure(figsize=(6,4))

    plt.scatter(X_test[column], y_test, color='blue', label='Actual')
    plt.scatter(X_test[column], y_pred, color='red', label='Predicted', alpha=0.5)

    plt.xlabel(column)
    plt.ylabel(y.name if y.name else 'Target')
    plt.title(f'{column} vs Target')
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
plt.figure(figsize=(6, 6))
plt.scatter(y_test, y_pred, color='purple', alpha=0.6)
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--', linewidth=2)
plt.xlabel("Actual Sales")
plt.ylabel("Predicted Sales")
plt.title("Actual vs Predicted (Multiple Linear Regression)")
plt.grid(True)
plt.show()
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

print("Mean Absolute Error (MAE):", mean_absolute_error(y_test, y_pred))
print("Mean Squared Error (MSE):", mean_squared_error(y_test, y_pred))
print("R² Score:", r2_score(y_test, y_pred))
```
